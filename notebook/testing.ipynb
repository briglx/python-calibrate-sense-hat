```python
def get_char_pixels(s):
        """
        Internal. Safeguards the character indexed dictionary for the
        show_message function below
        """

        if len(s) == 1 and s in text_dict.keys():
            return list(text_dict[s])
        else:
            return list(text_dict['?'])
```
